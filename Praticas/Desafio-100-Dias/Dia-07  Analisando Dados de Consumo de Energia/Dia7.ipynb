üìÇ Dataset: Informa√ß√µes sobre o consumo mensal de energia el√©trica em diferentes resid√™ncias, incluindo regi√£o, consumo em kWh e custo total.

üéØ Atividades:

Qual √© o consumo m√©dio de energia por regi√£o?

Existe correla√ß√£o entre o consumo em kWh e o custo total da conta?

Qual a resid√™ncia com o maior consumo registrado?

Crie um gr√°fico de linhas mostrando a varia√ß√£o do consumo m√©dio ao longo dos meses.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#testando dataset
df = pd.read_csv("dia_7_consumo_energia.csv")
df.head()
```

```python
#Qual √© o consumo m√©dio de energia por regi√£o?
consumoMedio = df.groupby("Regiao")["Consumo_kWh"].mean().sort_values(ascending=False)
print(f"Consumo m√©dio por regi√£o: \n\n{consumoMedio.apply(lambda x: f"{x:.1f} kWh").to_string()}")
```

```python
#Existe correla√ß√£o entre o consumo em kWh e o custo total da conta?
corrKWhCusto = df["Consumo_kWh"].corr(df["Custo_Total"])
print(f"Correla√ß√£o de: {corrKWhCusto:.2f}")
print("indica uma rela√ß√£o muito fraca entre as duas vari√°veis analisadas.")
```

```python
#Qual a resid√™ncia com o maior consumo registrado?
casaConsumo = df.groupby("ID_Residencia")["Consumo_kWh"].sum().sort_values(ascending=False)
print(f"A Resid√™ncia com maior consumo √© a: \n\n{casaConsumo.apply(lambda x: f'{x:.1f} kWh').head(1).to_string()}")
```

```python
#Crie um gr√°fico de linhas mostrando a varia√ß√£o do consumo m√©dio ao longo dos meses.
plt.figure(figsize=(12, 6))
plt.xticks(rotation=45)
sns.set_style("whitegrid")
sns.lineplot(data=df, x='Mes', y='Consumo_kWh', color='red')
plt.show()
```
