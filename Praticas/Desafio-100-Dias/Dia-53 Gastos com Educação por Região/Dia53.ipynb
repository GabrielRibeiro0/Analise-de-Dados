Este conjunto de dados apresenta os investimentos anuais em educa√ß√£o por regi√£o brasileira entre 2015 e 2023, incluindo n√∫mero de alunos e gasto per capita estimado.

üîç Tarefas:

    Qual foi o investimento total em educa√ß√£o (em milh√µes) por ano no Brasil?

    Qual regi√£o teve o maior gasto per capita m√©dio entre 2015 e 2023?

    Houve aumento no gasto per capita ao longo dos anos?

    Gere um gr√°fico de linha comparando o gasto per capita de cada regi√£o ao longo do tempo.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dia_53_gastos_educacao.csv")
df.head()
```

```python
#Qual foi o investimento total em educa√ß√£o (em milh√µes) por ano no Brasil?
investimentoEducacaoAno = df.groupby("Ano")["Investimento_Total_Milhoes"].sum()
print(investimentoEducacaoAno)
```

```python
#Qual regi√£o teve o maior gasto per capita m√©dio entre 2015 e 2023?
regiaoGastoPerCapita = df.groupby("Regi√£o")["Gasto_Per_Capita"].mean().round(2)
print(regiaoGastoPerCapita.idxmax())
```

```python
#Houve aumento no gasto per capita ao longo dos anos?
gastoPerCapitaAno = df.groupby("Ano")["Gasto_Per_Capita"].mean().round(2)
print(gastoPerCapitaAno)
```

```python
#Gere um gr√°fico de linha comparando o gasto per capita de cada regi√£o ao longo do tempo.

plt.figure(figsize=(10,6))
sns.lineplot(df, x="Ano", y="Gasto_Per_Capita", hue="Regi√£o", marker="o")
plt.title('Gasto per capita por Regi√£o (2015‚Äì2023)')
plt.xlabel('Ano')
plt.ylabel('Gasto per capita (R$)')
plt.legend(title='Regi√£o')
plt.grid()
plt.show()
```
